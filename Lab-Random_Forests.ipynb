### Instructions

- Apply the Random Forests algorithm but this time only by upscaling the data using `SMOTE`.
- Note that since `SMOTE` works on numerical data only, we will first encode the categorical variables in this case.


```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
numerical = pd.read_csv('C:/Users/aguva/Ironhack/Week_18/lab-random-forests/files_for_lab/numerical.csv')
categorical = pd.read_csv('C:/Users/aguva/Ironhack/Week_18/lab-random-forests/files_for_lab/categorical.csv')
targets = pd.read_csv('C:/Users/aguva/Ironhack/Week_18/lab-random-forests/files_for_lab/target.csv')
data = pd.concat([numerical, categorical, targets], axis = 1)
```

```python
data.head()
```

```python
encoder = OneHotEncoder(drop='first').fit(categorical)
encoded_categorical = encoder.transform(categorical).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical)
```

```python
data = pd.concat([numerical, encoded_categorical, targets], axis = 1)
```

```python
data.head()
```

```python
data.info()
```

```python
regression_target = data['TARGET_D']
# data.head()
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis = 1)

X.columns = X.columns.astype(str)
```

```python
data['TARGET_B'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_sm, y_sm = smote.fit_resample(X, y)
y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.25, random_state=0)
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
print(rf_classifier.score(X_test, y_test))
```
